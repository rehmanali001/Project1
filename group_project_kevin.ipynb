```python
# Dependencies
import requests
import json
import pandas as pd
```

```python
url = "http://api.zip-codes.com/ZipCodesAPI.svc/1.0/GetZipCodeDetails"
```

```python
zip_list =[]
pop_list =[]
lati_list =[]
long_list =[]

zip = [78702, 78703, 78704, 78717, 78721, 78722, 78723, 78724, 78726, 78727, 78728, 78729, 78730, 78731, 78732, 78733, 78734, 78735, 78737, 78738, 78739, 78741, 78744, 78745, 78746, 78748, 78749, 78750, 78751, 78752, 78753, 78754, 78757, 78758, 78759]
for z in zip:
    response = requests.get(url + f"/{z}?key=<L6SFRRWK211R3FST0O33>")
    response_json = response.json()
    response_zip = response_json["item"]["ZipCode"]
    zip_list.append(response_zip)
    response_pop = response_json["item"]["PopulationEstimate"]
    pop_list.append(response_pop)
    response_pop = response_json["item"]["Latitude"]
    lati_list.append(response_pop)
    response_pop = response_json["item"]["Longitude"]
    long_list.append(response_pop)
print(zip_list)
print(pop_list)
print(lati_list)
print(long_list)
```

```python
merged_pd = pd.DataFrame({'Zip Code': zip_list, '2017 Est.Population ':pop_list, 'Latitude': lati_list, 'Longitude': long_list})
merged_pd.head()
```

```python
median_sales_price_csv = "Median_Sale_Price_By_Zip.csv"
```

```python
median_sales_price_df = pd.read_csv(median_sales_price_csv)
median_sales_price_df.head()
```
