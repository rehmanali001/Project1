```python
# Save config information
url = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=1&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"
url2 = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=2&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"
url3 = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=3&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"
url4 = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=4&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"
url5 = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=5&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"
url6 = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=6&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"
url7 = "https://api.schooldigger.com/v1.1/schools?st=TX&city=Austin&page=7&perPage=50&appID=d560d9ed&appKey=09aa2483b03ad827e06c62326b969070"

```

```python
response = requests.get(url).json()
response2 = requests.get(url2).json()
response3 = requests.get(url3).json()
response4 = requests.get(url4).json()
response5 = requests.get(url5).json()
response6 = requests.get(url6).json()
response7 = requests.get(url7).json()
```

```python
final_response = response['schoolList'] + response2['schoolList'] + response3['schoolList'] + response4['schoolList'] + response5['schoolList'] + response6['schoolList'] + response7['schoolList']
final_response
```

```python
#creating a dataframe
final_response_df = pd.DataFrame(final_response)
final_response_df.head()
```

```python
final_response_df.count()
```

```python
#Build a dataframe for zipcode from the address dictionary
zipcode = []
address_df = final_response_df["address"]

for address in address_df:
    zipcode.append(address["zip"])
    
    #print(address["zip"])
zipcode
final_response_df['zip'] = zipcode

final_response_df
```

```python
#Build a dataframe for zipcode from the address dictionary
stars = []
rank_df = final_response_df["rankHistory"]

for rankHistory in rank_df:
    if (rankHistory == None):
        stars.append("none")
        #print("none")
    else:
        stars.append(rankHistory[0]["rankStars"])
        #print(rankHistory[0]["rankStars"])

final_response_df['stars'] = stars
final_response_df
```

```python
schools_df = final_response_df[["schoolid","schoolName","highGrade","zip","stars"]]
schools_df
```

```python
#export school list with zip code, rankings, and grade level to csv
schools_df.to_excel("school_rankings.xlsx", index=False)

```

```python
schools = schools_df[schools_df["stars"] != "none"]
schools = schools[schools["highGrade"] != "12"]
schools = schools[schools["highGrade"] != "11"]
schools = schools[schools["highGrade"] != "10"]
schools = schools[schools["highGrade"] != "9"]
schools = schools[schools["highGrade"] != "8"]
schools = schools[schools["highGrade"] != "7"]
schools = schools[schools["highGrade"] != "6"]
schools = schools[schools["highGrade"] != "PK"]
schools["stars"]=pd.to_numeric(schools["stars"])
schools
```

```python
schools.dtypes
```

```python
grouped_zip_df = schools.groupby(['zip'])
grouped_zip_df.head()
#print(grouped_zip_df)
```

```python
average_stars_df = grouped_zip_df.mean()
average_stars_df['zipcode']=average_stars_df.index
average_stars_df

#
```

```python
average_stars_df["stars"] = average_stars_df["stars"].map("{:.2f}".format)
average_stars_df

#
```

```python
average_stars_df.dtypes

```

```python
average_stars_df["stars"]=pd.to_numeric(average_stars_df["stars"])
average_stars_df

#
```

```python
average_stars_df.dtypes



```

```python
#Create a bar chart based off of the group series from before
average_stars_chart = average_stars_df.plot(kind='bar')
#set the xlabel and ylabel using class methods
average_stars_chart.set_xlabel("Zip Codes")
average_stars_chart.set_ylabel("Number of Stars (max 5)")

```

```python
# set x axis and tick locations
#avg_stars_df = average_stars_df.set_index("zip")
avg_stars_df = average_stars_df[["stars"]]
avg_stars_df.head()

x_axis = np.arange(len(avg_stars_df))
tick_locations = [value for value in x_axis]


```

```python
#create a list indicating where to write x lables and set figure size to adjust for space
plt.figure(figsize=(20,3))
plt.bar(x_axis, average_stars_df["stars"], color='b', alpha=.5, align="center")
plt.xticks(tick_locations, average_stars_df["zipcode"], rotation="vertical")
#Set a Title and labels
plt.title("Average Star Ratings by Zip Code for elementary schools")
plt.xlabel("Zip Code")
plt.ylabel("Star Rating")
plt.tight_layout()


```
